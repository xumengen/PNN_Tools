```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
plt.figure()  # create a figure window
x = np.array([0.05, 0.2, 0.25, 0.3, 0.4, 0.43, 0.48, 0.6, 0.7, 0.8, 0.9, 0.95])
y = x / 2
plt.plot(x, y, label='x/2')
g_x = [0.0863, 0.2662, 0.2362, 0.1687, 0.1260, 0.1756, 0.3290, 0.6694, 0.4573, 0.3320, 0.4063, 0.3535]
plt.plot(x, g_x, label='g(x)')
plt.xlim((0.05, 0.95))  # range of x
plt.ylim((0, 0.9))  # range of y
plt.xlabel('x')
plt.ylabel('y')
plt.legend()  # if g(x) >= x/2, class 1; otherwise, class 2
```
